```python
#DATA ENCODING
#CONVERTING CATEGORICAL FEATURES INTO NUMERICAL
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
```

```python
emp_data['Department'] = label_encoder.fit_transform(emp_data['Department'])
emp_data['EducationField'] = label_encoder.fit_transform(emp_data['EducationField'])
emp_data['JobRole'] = label_encoder.fit_transform(emp_data['JobRole'])
emp_data['MaritalStatus'] = label_encoder.fit_transform(emp_data['MaritalStatus'])
emp_data['Attrition'] = label_encoder.fit_transform(emp_data['Attrition'])
emp_data['BusinessTravel'] = label_encoder.fit_transform(emp_data['BusinessTravel'])
emp_data['Gender'] = label_encoder.fit_transform(emp_data['Gender'])
emp_data['Over18'] = label_encoder.fit_transform(emp_data['Over18'])
emp_data['OverTime'] = label_encoder.fit_transform(emp_data['OverTime'])
```

```python
emp_data.head()
```

```python
#DEPENDENT(TARGET) AND INDEPENDENT VARIABLES:
x=emp_data.drop('Attrition',axis=1)
y=emp_data['Attrition']
```

```python
# SPLITTING DATA INTO TRAINING AND TESTING
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_train.head()
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
std_scaler= StandardScaler()
```

```python
xtrain_scaled=std_scaler.fit_transform(x_train)
xtest_scaled=std_scaler.transform(x_test)
```

```python
xtrain_scaled
xtest_scaled
```

```python
#MODEL DEVOLOPMENT
#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
Random_forest_model=RandomForestClassifier()
```

```python
#MODEL TRAINING
Random_forest_model.fit(xtrain_scaled,y_train)
```

```python
#MODEL PREDICTION
y_pred=Random_forest_model.predict(xtest_scaled)
y_pred
```

```python
#MODEL EVALUATION
#CONFUSION MATRIX
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt='d')
```

```python
#ACCURACY
from sklearn.metrics import accuracy_score
model_accuracy=accuracy_score(y_test,y_pred)
print("Acuracy of the model=",model_accuracy)
```

```python
#PRECISION
from sklearn.metrics import precision_score
model_precision=precision_score(y_test,y_pred)
print("precision of the model=",model_precision)
```

```python
#RECALL
from sklearn.metrics import recall_score
model_recall=recall_score(y_test,y_pred)
print("recall of the model=",model_recall)
```

```python
#F1 SCORE
from sklearn.metrics import f1_score
f1score=f1_score(y_test,y_pred)
print("f1_score of the model=", f1score)
```

```python
#CLASSIFICATION REPORT
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#FEATURES IMPORTANCE METRIC IN RANDOM FOREST/ FEATURE EXTRACTION
score_list=Random_forest_model.feature_importances_
list_of_features=list(x.columns)
score_df=pd.DataFrame({"feature":list_of_features,'score':score_list})
score_df.sort_values(by='score',ascending=False)
```

```python
list_of_features = list(x.columns)
plt.figure(figsize=(2, 8))
plt.barh(range(len(list_of_features)), Random_forest_model.feature_importances_)  # Corrected attribute name
plt.yticks(np.arange(len(list_of_features)), list_of_features)  # Corrected method name
plt.ylabel('Features')
plt.xlabel('Importance')
plt.show()
```

```python
# K-FOLD CROSS VALIDATIONS
from sklearn.model_selection import cross_val_score
scores=cross_val_score(Random_forest_model,xtrain_scaled,y_train,cv=5,scoring='accuracy')
print('cross-validation scores',scores)
```

```python
Avg_model_score=scores.mean()
print('Average model score',Avg_model_score)
```
